# Residence time distributions from CSTR and Van de Vusse

Compare RTDs using series CSTR and Vusse approach using functions available in the rtd.py module of PyroPy.

```python
import numpy as np
import matplotlib.pyplot as py
%matplotlib inline
import chemics as cm
    
# Compare Series CSTR and Vusse RTD
# -----------------------------------------------------------------------------

n = 8                       # number of stages
tau = 2.8                   # residence time, s
t = np.linspace(0, 10, 200) # time vector, s

rt_scstr = cm.scstr(n, tau, t)          # RTD based on series CSTR
rt_circ = cm.vusse(n, 2.8, t)           # RTD based on Vusse 1962, with circulation
rt_nocirc = cm.vusse(n, 2.8, t, r=1e-6) # RTD based on Vusse 1962, no circulation

# Plot
# -----------------------------------------------------------------------------

py.ion()
py.close('all')

py.figure(1)
py.plot(t, rt_scstr, 'b-', lw=2, label='series CSTR')
py.plot(t, abs(rt_circ.real), 'g-', lw=2, label='Vusse, q/r=1')
py.plot(t, abs(rt_nocirc.real), 'r--', lw=2, label='Vusse, r=0')
py.legend(loc='best', numpoints=1)
py.ylabel('RTD')
py.xlabel('Time (s)')
py.grid()
```
